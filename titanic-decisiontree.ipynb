```python
df=pd.read_csv('/kaggle/input/titanic/train.csv')
df
```

```python
df.info()
```

```python
titanic_train=df.copy()
titanic_train
```

```python
titanic_train.drop("Cabin",axis=1,inplace=True)
titanic_train.drop("Name",axis=1,inplace=True)
titanic_train.drop("Ticket",axis=1,inplace=True)
```

```python
titanic_train.info()
```

```python
medians = titanic_train["Age"].median()
titanic_train["Age"]=titanic_train["Age"].fillna(medians)
medians1 = titanic_train["Fare"].median()
titanic_train["Fare"]=titanic_train["Fare"].fillna(medians1)
```

```python
one_hot_encoded_training_predictors = pd.get_dummies(titanic_train)
one_hot_encoded_training_predictors
```

```python
one_hot_encoded_training_predictors.info()
```

```python
one_hot_encoded_training_predictors['Fare'] = one_hot_encoded_training_predictors['Fare'].astype(int)
one_hot_encoded_training_predictors['Age'] = one_hot_encoded_training_predictors['Age'].astype(int)
```

```python
# Identify the boolean columns
boolean_columns = one_hot_encoded_training_predictors.select_dtypes(include=['bool']).columns
```

```python
for column in boolean_columns:
    one_hot_encoded_training_predictors[column] = one_hot_encoded_training_predictors[column].astype(int)
one_hot_encoded_training_predictors
```

```python
from sklearn.tree import DecisionTreeClassifier
x_train = one_hot_encoded_training_predictors.drop(['Survived'], axis=1)
y_train = one_hot_encoded_training_predictors['Survived']
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
```

```python
df1=pd.read_csv('/kaggle/input/titanic/test.csv')
df1
```

```python
df1.info()
```

```python
df1.drop("Ticket",axis=1,inplace=True)
df1.drop("Cabin",axis=1,inplace=True)
df1.drop("Name",axis=1,inplace=True)
```

```python
medians = df1["Age"].median()
df1["Age"]=df1["Age"].fillna(medians)
medians1 = df1["Fare"].median()
df1["Fare"]=df1["Fare"].fillna(medians1)
```

```python
df1
```

```python
df1['Fare'] = df1['Fare'].astype(int)
df1['Age'] = df1['Age'].astype(int)
```

```python
one_hot=pd.get_dummies(df1)
one_hot
```

```python
boolean_columns1 = one_hot.select_dtypes(include=['bool']).columns
```

```python
for column in boolean_columns1:
    one_hot[column] = one_hot[column].astype(int)
one_hot
```

```python
one_hot.info()
```

```python
#prediction of test csv files
error=clf.predict(one_hot)
error
```
